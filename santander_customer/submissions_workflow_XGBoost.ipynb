# XGBoost starter

```python
import xgboost as xgb
```

```python
X = train_df.iloc[:,2:]
y = train_df.target
```

```python
# Create the DMatrix
xgb_dmatrix = xgb.DMatrix(X, y)

# Create the parameter dictionary
params = {"objective":"binary:logistic"}
```

```python
# Train the model
xg_bin = xgb.train(params = params, dtrain = xgb_dmatrix)
```

```python
# Plotting feature importance
xgb.plot_importance(xg_bin)
```

```python
xgb.plot_importance(xg_bin, importance_type = 'gain')
```

```python
# https://xgboost.readthedocs.io/en/latest/python/python_api.html
xg_bin.get_score()
```

```python
xg_bin.get_score(importance_type = 'gain')
```

```python
X_test_df = pd.read_csv('test.csv')
features_X_test_df = X_test_df.iloc[:,1:]
```

```python
X_test_matrix = xgb.DMatrix(features_X_test_df)
```

```python
preds = xg_bin.predict(X_test_matrix)
preds.max()
```

```python
toy_submission = pd.DataFrame.from_dict({'ID_code': X_test_df['ID_code'],
                                         'target':  preds})
```

```python
toy_submission.to_csv('toy_submission.csv', index = False)
```

# Using grid search


```python
from sklearn.model_selection import GridSearchCV
```

Nice link about all parameters of XGBoost: https://www.analyticsvidhya.com/blog/2016/03/complete-guide-parameter-tuning-xgboost-with-codes-python/

Note that some parameters names changes from skiti-learn and xgboost. For example `eta` -> `learning_rate`

```python
gbm_hyperparam_grid = {'learning_rate': [0.01, 0.1, 0.2],
                       'n_estimators': [200],
                       'lambda': [0.8, 1],    # default: 1
                       'reg_alpha': [0, 0.1], # default: 0
                       'subsample': [0.3, 0.5],
                       'max_depth': [4, 5, 6],
                       'num_boosting_rounds': [10, 20]}
```

```python
gbm = xgb.XGBRegressor()
grid_mse = GridSearchCV(estimator = gbm,
                        param_grid = gbm_hyperparam_grid, 
                        scoring = 'neg_mean_squared_error', 
                        cv = 4,      # 4-fold crossvalidation
                        verbose = 3) # Prints the grid search profile

# Fit the grid to the data
grid_mse.fit(X, y)

best_params = grid_mse.best_params_
best_params["objective"] = "binary:logistic"

# Create the DMatrix
xgb_dmatrix = xgb.DMatrix(X, y)

# Train the model from the best parameters of the grid search
best_xg_reg = xgb.train(params = best_params, dtrain = xgb_dmatrix)
```

# This took around 4 days to run and gave the worts results (This was because I didn't parallelize with n_jobs = -1 inside the GridSearchCV!). Even worst than the XGBoost default parameters.



```python
import pickle
# Save results
with open('best_xg_reg.pkl', 'wb') as output:
    pickle.dump(best_xg_reg, output, pickle.HIGHEST_PROTOCOL)
```

```python
preds_best = best_xg_reg.predict(X_test_matrix)
preds_best.max()
```

```python
toy_submission_best = pd.DataFrame.from_dict({'ID_code': X_test_df['ID_code'],
                                              'target':  preds_best})
```
